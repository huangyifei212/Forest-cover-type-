# 偏度分析  
在概率论和统计学中，偏度衡量实数随机变量概率分布的不对称性。偏度的值可以为正，可以为负或者甚至是无法定义。在数量上，偏度为负（负偏态）就意味着在概率密度函数左侧的尾部比右侧的长，绝大多数的值（不一定包括中位数在）位于平均值的右侧。偏度为正（正偏态）就意味着在概率密度函数右侧的尾部比左侧的长，绝大多数的值（不一定包括中位数）位于平均值的左侧。偏度为零就表示数值相对均匀地分布在平均值的两侧，但不一定意味着其为对称分布。

```python
soil_counts = []
for num in range(1,41):
    col = ('soil_type_' + str(num))
    this_soil = data[col].groupby(data['cover_type'])
    totals = []
    for value in this_soil.sum():
        totals.append(value)
    total_sum = sum(totals)
    soil_counts.append(total_sum)
    print("Total Trees in Soil Type {0}: {1}".format(num, total_sum))
    percentages = [ (total*100 / total_sum) for total in totals]
    print("{0}\n".format(percentages))
print("Number of trees in each soil type:\n{0}".format(soil_counts))
```

```python
#考虑的特征变量
size=10
datatp=data.iloc[:,:size]
#获得列名称
cols=datatp.columns
#计算相关系数
data_corr=datatp.corr()
#设定阈值选择较高值的属性
thr=0.5
corr_list=[]
#搜索高度相关的对
for i in range(0,size):  
    for j in range(i+1,size):  
        if (data_corr.iloc[i,j] >= thr  and data_corr.iloc[i,j] < 1) or (data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -thr):
            corr_list.append([data_corr.iloc[i,j],i,j])  

#排序            
s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))

#展示几对相关性较强的对
for v,i,j in s_corr_list:
    print ("corr(%s and %s )= %.2f" % (cols[i],cols[j],v))
```

```python
X=data.loc[:,'elevation':'soil_type_40']
y=data['cover_type']
```

```python
rm=['hillshade_3pm','soil_type_7','soil_type_8','soil_type_14','soil_type_15',
     'soil_type_21','soil_type_25','soil_type_28','soil_type_36','soil_type_37']
#选择要删除的属性特征
```

```python
X.drop(rm, axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=123)
```

```python
%%time

neighbors = np.arange(1,7)
train_accuracy =np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test) 
```

```python
plt.figure(figsize=(10,6))
plt.title('Sensitivity analysis of the number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy',marker='*',color='b')
plt.plot(neighbors, train_accuracy, label='Training accuracy',marker='o',color='r')
plt.legend()
for a, b in zip(neighbors, test_accuracy):
    plt.text(a, b, "%.5f"%b, ha='center', va='bottom', fontsize=10)
for a, b in zip(neighbors, train_accuracy):
    plt.text(a, b,"%.5f"%b, ha='center', va='bottom', fontsize=10)
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')

plt.savefig("Sensitivity analysis of the number of neighbors.png")
```
